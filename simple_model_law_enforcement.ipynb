# Illustration de l'effet dissuasif de $S$ et $q$ à l'aide des graphiques

```python
from sympy import *
from matplotlib import pyplot as plt
import numpy as np
import matplotlib
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['text.latex.preamble'] = [r'\usepackage{amsmath}']
init_printing(use_unicode=True)
```

```python
# Définition des symboles
q, b, S, h, w, B, h0 = symbols('q, b, S, h, w, bbar, h0', positive=True)
f = symbols('f', cls=Function)(b)
C = symbols('C', cls=Function)(q)
```

## fonction de densité croissante $f(b) = \exp(-b)$ et coût quadratique $C(q) = q^2$

```python
# Définition de la fonction de densité de probabilité
f = exp(-b) # fonction de densité exponentielle

# Définition du coût C
C = q**2 # coût quadratique
```

```python
# Définition de la fonction de bien-être social
funct = (b - h) * f
welfare = integrate(funct, (b, q*S, B)) - C
welfare
```

```python
# Définition de la valeur des paramètres
B_val = 10
h0_val = 1
h_val = 5
q_val = 0.5
S_val = 2
```

```python
# Tracé de la fonction de bien-être social W en fonction de q pour différentes valeurs de S
q_range = np.linspace(0, 1, 100)
S_values = [1, 3, 5]

W_values = []
for S_val in S_values:
    welfare_q = welfare.subs([(B, B_val), (h, h_val), (h0, h0_val), (S, S_val)])
    W_values.append([float(welfare_q.subs(q, q_v)) for q_v in q_range])

plt.figure(figsize=(8, 6))
for i in range(len(S_values)):
    plt.plot(q_range, W_values[i], label='$S = {}$'.format(S_values[i]))
plt.xlabel(r'$q$')
plt.ylabel(r'$W$')
plt.title(r'$W(q,S_{i}), f(b)=\exp(-b) $', fontsize=12)
plt.legend()
plt.savefig('Dissuasion_q_fexpo', dpi=300)
plt.show()
```

```python
# Tracé de la fonction de bien-être social W en fonction de S pour différentes valeurs de q
S_range = np.linspace(0, 5, 100)
q_values = [0.1, 0.5, 0.9]

W_values = []
for q_val in q_values:
    welfare_S = welfare.subs([(B, B_val), (h, h_val), (h0, h0_val), (q, q_val)])
    W_values.append([float(welfare_S.subs(S, S_v)) for S_v in S_range])

plt.figure(figsize=(8, 6))
for i in range(len(q_values)):
    plt.plot(S_range, W_values[i], label='$q = {}$'.format(q_values[i]))
plt.xlabel(r'$S$')
plt.ylabel(r'$W$')
plt.title(r'$ W(q_{j},S), f(b)= \exp(-b) $', fontsize=12)
plt.legend()
plt.savefig('Dissuasion_S_fexpo', dpi=300)
plt.show()
```
