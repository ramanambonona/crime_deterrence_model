# Illustration des utilités espérées en fonction de $q$, $b$ et $S$

```python
from __future__ import unicode_literals
import numpy as np
import matplotlib
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['text.latex.preamble'] = [r'\usepackage{amsmath}']
import matplotlib.pyplot as plt

%matplotlib inline
```

## EU ($s$, $q$ élevé), exemple $U(b) = b^2$

```python
# fonction d'utilité convexe
U = lambda b: b**2
q = 0.9
b = np.linspace(0, 100, 100)

# domaine de définition pour S
S = np.arange(0, 100, 20)

# Tracer les courbes
fig, ax = plt.subplots()

for i in range(len(S)):
    UE = (1-q)*U(b) - q*U(b-S[i])
    ax.plot(b, UE, label='$S = $'+str(S[i]))

ax.set_xlabel(r'$b$')
ax.set_ylabel(r'$UE$')
ax.set_title(r'$ \textit{EU en fonction de b  q =0.9}$', fontsize=12, color='black')
ax.legend()
plt.show()
plt.savefig('EU_U_quadratic_q_0.9.png', dpi=300)
```

## probilité $q$ fonction inverse de b

```python
def U(b):
    return b

# Fonction q en fonction inverse du b
def q(b, S):
    return S/b

# Définition de UE(b)
def UE(b, S):
    return (1-q(b, S))*U(b) - q(b, S)*U(b-S)

# Tracer UE pour différentes valeurs de S
b = np.arange(0, 100, 20)
for S in [0, 25, 50, 75, 100]:
    UE_values = UE(b, S)
    plt.plot(b, UE_values, label=r'$S = $'+str(S))

plt.xlabel(r'$b$')
plt.ylabel(r'$UE$')
plt.title('Effet de S, p inverse')
plt.legend()
plt.show()

plt.savefig('EU_q_inverse.png', dpi=300)
```

## Aversion au risque, fonction d'utilité CRRA, mesure d'aversion $\rho$

### CRRA $U(b)= \frac{b^{1-\rho}}{1-\rho}$ pour  $\rho = 0$

```python
# AVERSION AU RISQUE AVEC CRRA courbe
def CRRA1(b, rho):
    return b**(1-rho)/(1-rho)

def CRRA2(b, s, rho):
    return (b-s)**(1-rho)/(1-rho)

def UE(b, q, s, rho):
    return (1-q)*CRRA1(b, rho) - q*CRRA2(b, s, rho)
# UE(b,S)= (1-p)\frac{b^{1-\rho}}{1-\rho}+p(1-p)\frac{(b-S)^{1-\rho}}{1-\rho}
b = np.linspace(0, 20, 100)
rho = 0
s1, s2 = 10, 20
q1, q2 = 0.1, 0.9

ue_1 = UE(b, q1, s1, rho)
ue_2 = UE(b, q2, s1, rho)
ue_3 = UE(b, q1, s2, rho)
ue_4 = UE(b, q2, s2, rho)

plt.plot(b, ue_1, label=r'$U_1 $ pour $S={}$ et $q={}$'.format(s1, q1))
plt.plot(b, ue_2, label=r'$U_2 $ pour $S={}$ et $q={}$'.format(s1, q2))
plt.plot(b, ue_3, label=r'$U_3 $ pour $S={}$ et $q={}$'.format(s2, q1), linestyle='--')
plt.plot(b, ue_4, label=r'$U_4$ pour $S={}$ et $q={}$'.format(s2, q2), linestyle='--')

plt.xlabel(r'$b$')
plt.ylabel(r'$UE$')
plt.legend()
plt.title(r"CRRA $\rho = 0$")
plt.savefig('crra_rho_0.png', dpi=300)
plt.show()
```

# Effet de la probabilité $q$ sur le bien être $W$, coût quadratique et $f$ uniforme

```python
h = 1
S = 2
q = np.linspace(0, 1, 100)
C = q**2
f = np.ones(100)
b = np.linspace(0, 5, 100)

# Calculer W
W = np.zeros(100)
for i in range(100):
    W[i] = np.sum((b[b >= q[i]*S] - h)*f[b >= q[i]*S]) - C[i]

# Tracer le graphique
fig, ax = plt.subplots()
ax.grid()
ax.plot(q, W, color='black')
ax.set_xlabel(r'$q$')
ax.set_ylabel(r'$W$')
ax.set_title(r'$W(q, \bar{S})$ avec $f$ uniforme')
plt.savefig('Welfare_f_uniforme.png', dpi=300)
plt.show()
```

# Effet de la sanction $S$ sur le bien être $W$, coût quadratique, avec $f(b) = \exp(-b)$, Recheche d'une sanction optimale 

```python
def f(b):
    return np.exp(-b)

def C(q):
    return q**2

h = 5
q = np.linspace(0, 1, 100)
C = q**2
f = np.ones(100)
b = np.linspace(0, 5, 100)

S_values = np.linspace(0, 5, 100)
W_values = np.zeros(100)

for i, S in enumerate(S_values):
    W = np.zeros(100)
    for j in range(100):
        W[j] = np.sum((b[b >= q[j]*S] - h)*f[b >= q[j]*S]) - C[j]
    W_values[i] = np.max(W)

# Trouver le S optimal
S_optimal = S_values[np.argmax(W_values)]
print("S optimal:", S_optimal)

# Tracer le graphique
fig, ax = plt.subplots()
ax.grid()
ax.plot(S_values, W_values, color='black')
ax.set_xlabel(r'$S$')
ax.set_ylabel(r'$W$')
ax.set_title(r'$W(\bar{q}, S)$ avec $f(b) =\exp(-b)$')
plt.savefig('Welf_S_max_exp.png', dpi=300)
plt.show()
```

# Effet de la probabilité $q$ sur le bien être $W$, coût quadratique, avec $f(b) = \exp(-b)$, Recheche d'une probabilité optimale 

```python
def f(b):
    return np.exp(-b)

def C(q):
    return q**2

b = np.linspace(0, 5, num=100)
h = 5

# Trouver S optimal
def find_S_optimal(h, q):
    S_range = np.linspace(0, 10, num=100)
    W = np.zeros(100)
    for i, S in enumerate(S_range):
        W[i] = np.sum((b[b >= q*S] - h)*f(b[b >= q*S])) - C(q)
    return S_range[np.argmax(W)]

# Trouver p optimal
q_range = np.linspace(0, 1, num=100)
W = np.zeros(100)
for i, q in enumerate(q_range):
    S = find_S_optimal(h, q)
    W[i] = np.sum((b[b >= q*S] - h)*f(b[b >= q*S])) - C(q)
q_optimal = q_range[np.argmax(W)]

print("q optimal:", q_optimal)
# Tracer le graphique
fig, ax = plt.subplots()
ax.plot(q_range, W, color='black')
ax.set_xlabel(r'$q$')
ax.set_ylabel(r'$W$')
ax.set_title(r'$W(q, \bar{S})$ avec $f(b) =\exp(-b)$')
plt.savefig('Welf_q_opt_exp.png', dpi=300)
plt.show()
```

# Effet de la probabilité $q$ sur le bien être $W$, coût quadratique, avec $f(b) = \exp(-b)$

```python
def f(b):
    return np.exp(-b)

def C(q):
    return q**2

b = np.linspace(0, 5, num=100)
h = 2
S = 4

q = np.linspace(0, 1, num=100)
W = np.zeros(100)

for i in range(100):
    W[i] = np.sum(f(b[b >= q[i]*S]) * (b[b >= q[i]*S] - h)) - C(q[i])

fig, ax = plt.subplots()
ax.plot(q, W, color='black')
ax.set(xlabel=r'$q$', ylabel=r'$W$',
       title= r'$W(q, \bar{S})$ avec $f(b) =\exp(-b)$')
ax.grid()
plt.savefig('Welf_h_seuil_exp.png', dpi=300)
plt.show()
```

```python
def f(b):
    return 1 / (1 + np.exp(-b))

def C(q):
    return q**2

b = np.linspace(0, 5, num=100)
h = 2
S = 4

q = np.linspace(0, 1, num=100)
W = np.zeros(100)

for i in range(100):
    W[i] = np.sum(f(b[b >= q[i]*S]) * (b[b >= q[i]*S] - h)) - C(q[i])

fig, ax = plt.subplots()
ax.plot(q, W, color='black')
ax.set(xlabel=r'$q$', ylabel=r'$W$',
       title=r'$W(q, \bar{S})$ avec $f(b) =\frac{1}{1 + \exp(-b)}$')
ax.grid()
plt.savefig('Welf_f_logistique.png', dpi=300)

plt.show()
```
