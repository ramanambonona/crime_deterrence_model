# Interprétation

L'analyse qu'on peut tirer à partir de la courbe obtenue est que la dissuasion du crime dépend de la probabilité de détection q, de la sanction maximale S et du bénéfice du crime b. La fonction d'utilité choisie dans ce code est $U = b^2$, qui est une fonction d'utilité convexe.

Pour chaque valeur de $S$, la courbe représente l'espérance d'utilité pour un criminel qui prend en compte le coût de la détection ($q * U(b-S)$) et le gain du crime ($U(b)$). On peut voir que pour une probabilité de détection fixe ($q = 0,9$), l'espérance d'utilité augmente avec le bénéfice du crime $b$ jusqu'à atteindre une valeur maximale, puis décroît. 
Lorsque la courbe décroît à un certain niveau, cela signifie que le bénéfice obtenu par le criminel est inférieur à ce qu'il aurait pu être s'il n'y avait pas de sanctions. En d'autres termes, la décroissance de la courbe indique que la dissuasion du crime fonctionne et que les sanctions ont un impact sur les décisions criminelles.

Cette décroissance est représente l'effet de dissuasion. Elle est générée(pour une probabilité fixée) par l'augmentation de la severité de la sanction, ce qui réduit le bénéfice net obtenu par le criminel. En conséquence, le criminel est moins incité à commettre le crime et la société est moins affectée par les coûts associés au crime.

La courbe décroît à un certain niveau en raison de l'effet combiné de la probabilité de détection $q=0.9$ et de la severité de la sanction $S$. En augmentant la severité de la sanction $S$, la courbe décroît également plus rapidement, ce qui montre que les criminels sont de moins en moins incités à commettre le crime.

Lorsque la sanction maximale $S$ augmente, l'espérance d'utilité pour le criminel diminue pour toutes les valeurs de b, ce qui signifie que le risque d'être pris et puni est plus élevé, ce qui dissuade le criminel de commettre le crime. Cependant, pour une valeur suffisamment élevée de bénéfice du crime, le criminel peut être toujours incité à commettre le crime même si la sanction maximale est élevée.

```python
# fonction d'utilité convexe
U = lambda b: b**2
q = 0.1
b = np.linspace(0, 100, 100)

# domaine de définition pour S
S = np.arange(0, 100, 20)

# Tracer les courbes
fig, ax = plt.subplots()

for i in range(len(S)):
    UE = (1-q)*U(b) - q*U(b-S[i])
    ax.plot(b, UE, label='S = '+str(S[i]))

ax.set_xlabel(r'$b$')
ax.set_ylabel(r'$UE$')
ax.set_title(r'$ \textit{EU en fonction de b  q =0.1}$', fontsize=12, color='black')
ax.legend()
plt.show()
plt.savefig('EU_U_quadratic_q_0.1.png', dpi=300)
```

```python
# Fonction d'utilité U(b) convexe
def U(b):
    return b

# Fonction p(b) en fonction inverse du b
def q(b, S):
    return S/b

# Définition de UE(b)
def UE(b, S):
    return (1-q(b, S))*U(b) - q(b, S)*U(b-S)

# Tracer UE pour différentes valeurs de S
b = np.arange(0, 100, 20)
for S in [0, 25, 50, 75, 100]:
    UE_values = UE(b, S)
    plt.plot(b, UE_values, label=r'$S = $'+str(S))

plt.xlabel(r'$b$')
plt.ylabel(r'$UE$')
plt.title('Effet de S')
plt.legend()
plt.show()

plt.savefig('EU_q_variable.png', dpi=300)
```

L'expression $q = \frac{S}{b}$ définit la probabilité de détection comme étant directement proportionnelle au rapport entre la sanction maximale ($S$) et le bénéfice potentiel du crime ($b$).

Cela signifie que, plus le bénéfice potentiel est élevé, moins la probabilité de détection doit être élevée pour équilibrer les coûts et les bénéfices de la commission du crime. D'autre part, si la sanction maximale est plus élevée, la probabilité de détection nécessaire pour équilibrer les coûts et les bénéfices du crime doit également augmenter.

Lorsque nous observons la courbe UE en fonction de b pour différentes valeurs de $S$, nous pouvons voir comment l'augmentation de $S$ influence l'analyse de la dissuasion. Plus la sanction maximale est élevée, plus la courbe UE baisse, ce qui signifie que la commission du crime devient moins avantageuse pour le criminel.

En d'autres termes, la dissuasion du crime est plus efficace pour des sanctions plus élevées. Cependant, il est également important de noter que si la sanction est fixe, l'efficacité de la dissuasion dépend du bénéfice potentiel du crime. Si le bénéfice potentiel est très élevé, une sanction élevée peut ne pas être suffisante pour dissuader le criminel de commettre le crime.

# intrepretation
La fonction d'utilité CRRA (fonction de la forme de la courbe de l'utilité qui décroît plus lentement ou plus rapidement) est utilisée ici pour représenter la décision de l'individu pour mesure son audace de commettre ou non le crime. Plus précisément, le modèle d'utilité espérée (UE) est défini  en utilisant la fonction CRRA.

Rappelons que l'UE pour un individu est égale à la somme pondérée de l'utilité potentielle du bénéfice du crime (b) et de l'utilité potentielle de la sanction (s) en cas de commission du crime, avec des poids donnés par la probabilité (q) de se faire prendre. L'UE est maximisée lorsque l'individu décide de ne pas commettre de crime s'il estime que le coût potentiel de la sanction est supérieur au bénéfice potentiel du crime.

Le graphique affiche quatre courbes d'UE pour différents niveaux de sanctions (s1 et s2) et différentes probabilités (q1 et q2) de se faire prendre. Les courbes concaves représentent les UE pour un niveau de sanction inférieur (s1) tandis que les courbes convexes représentent les UE pour un niveau de sanction plus élevé (s2). De même, les courbes solidaires représentent les UE pour une probabilité plus faible (q1) de se faire prendre, tandis que les courbes en pointillés représentent les UE pour une probabilité plus élevée (q2) de se faire prendre.

En d'autres termes, les courbes qui ne sont pas en pointillés représentent l'utilité espérée pour un niveau de sanction plus faible (s1) tandis que les courbes en pointillés représentent l'utilité espérée pour un niveau de sanction plus élevé (s2). Les courbes de couleur pleine représentent l'utilité espérée pour une probabilité plus faible de se faire prendre (q1), tandis que les courbes avec un tracé plus épais représentent l'utilité espérée pour une probabilité plus élevée de se faire prendre (q2).

Si la probabilité de se faire prendre est faible et la sanction est faible, l'utilité espérée du crime est plus élevée, ce qui peut inciter l'individu à commettre le crime. Si la probabilité de se faire prendre est plus élevée et/ou la sanction est plus élevée, l'utilité espérée du crime est plus faible, ce qui peut dissuader l'individu de commettre le crime. Ainsi, la courbe montre comment la dissuasion du crime dépend de la probabilité de se faire prendre et de la taille de la sanction.

L'analyse de ces courbes peut aider à comprendre les implications économiques d'une politique de dissuasion. Par exemple, une augmentation de la sanction (passant de s1 à s2) réduit l'utilité espérée du crime et peut inciter les individus à ne pas commettre de crimes. De même, une augmentation de la probabilité de se faire prendre (passant de q1 à q2) réduit également l'utilité espérée du crime. Les politiques de dissuasion telles que l'augmentation des sanctions et l'amélioration de la capacité de détection peuvent ainsi être efficaces pour réduire la criminalité.

La fonction d'utilité CRRA (Constant Relative Risk Aversion) est utilisée pour mesurer l'aversion au risque d'un individu. Ce modèle suppose que l'individu préfère un bénéfice constante plutôt qu'un bénéfice fluctuant, ce qui signifie que plus l'aversion au risque est forte, plus l'individu préfère un bénéfice stable à un bénéfice potentiellement plus élevé mais avec plus de risque.

Dans l'analyse de la dissuasion du crime, nous pouvons intégrer la fonction d'utilité CRRA pour mesurer les implications économiques de la politique de dissuasion en fonction du paramètre d'aversion au risque $\rho$. Si $\rho$ est petit, cela signifie que l'individu a une faible aversion au risque et est plus enclin à accepter un bénéfice potentiellement plus élevé même si elle est associée à un plus grand risque de se faire prendre et d'être sanctionné. Dans ce cas, la politique de dissuasion devra être plus forte pour dissuader l'individu de commettre le crime. À l'inverse, si $\rho$ est élevé, cela signifie que l'individu a une forte aversion au risque et préfère un bénéfice plus stable à un bénéfice potentiellement plus élevé mais avec plus de risque. Dans ce cas, la politique de dissuasion n'a pas besoin d'être aussi forte pour dissuader l'individu de commettre le crime.

En général, l'analyse de la dissuasion du crime montre que plus la sanction associée au crime est élevée et la probabilité d'être pris plus importante, plus l'utilité espérée associée au crime sera faible, ce qui dissuadera l'individu de le commettre. L'ajout de la fonction d'utilité CRRA dans l'analyse montre que ces deux facteurs peuvent être influencés par l'aversion au risque de l'individu, ce qui peut avoir des implications sur la politique de dissuasion nécessaire pour dissuader l'individu de commettre le crime.

Lorsque la probabilité de détection est faible ($q = 0,1$), l'utilité escomptée reste relativement stable et dépend peu de la valeur de montant de la peine ($s$). Cependant, lorsque la probabilité de détection est élevée ($q = 0,9$), l'utilité escomptée dépend de manière significative de la valeur de montant de la peine ($s$). Si la valeur de montant de la peine est élevée ($s = 20$), l'utilité escomptée est relativement faible pour toutes les valeurs de bénéfice, ce qui suggère que la dissuasion est plus forte.

Ainsi, cette courbe montre comment l'utilité escomptée dépend de la valeur de "probabilité de détection" et de "montant de la peine", ce qui peut aider à comprendre l'effet de la dissuasion sur les comportements criminels.

```python
def f(b):
    return np.exp(-b)

def C(q):
    return q**2

h = 5
q = np.linspace(0, 1, 100)
C = q**2
f = np.ones(100)
b = np.linspace(0, 5, 100)

S_values = np.linspace(0, 5, 100)
W_values = np.zeros(100)

for i, S in enumerate(S_values):
    W = np.zeros(100)
    for j in range(100):
        W[j] = np.sum((b[b >= q[j]*S] - h)*f[b >= q[j]*S]) - C[j]
    W_values[i] = np.max(W)

# Trouver le S optimal
S_optimal = S_values[np.argmax(W_values)]
print("S optimal:", S_optimal)

# Tracer le graphique
fig, ax = plt.subplots()
ax.plot(S_values, W_values, color='black')
ax.set_xlabel(r'$S$')
ax.set_ylabel(r'$W$')
ax.set_title(r'$W(\bar{q}, S)$ avec $f(b) =\exp(-b)$')
plt.savefig('Welf_S_max_exp.png', dpi=300)
plt.show()
```

Le but est de maximiser la différence entre les bénéfices potentiels des criminels et les coûts associés à la détection, ce qui se reflète dans la fonction $W(q, S)$. Le niveau optimal de détection se trouve en trouvant la probabilité q optimale qui maximise $W(q, S)$.

L'interprétation économique de la courbe obtenue est la suivante: plus la probabilité de détection est élevée, plus les coûts associés à la détection augmentent, ce qui réduit la valeur de W. Cependant, si la probabilité de détection est trop faible, les criminels peuvent tirer un bénéfice important sans être détectés, ce qui réduit également la valeur de W.

Le point optimal sur la courbe représente le compromis entre ces deux facteurs, offrant le niveau optimal de détection qui maximise la différence entre les bénéfices potentiels des criminels et les coûts associés à la détection.

```python
def f(b):
    return np.exp(-b)

def C(q):
    return q**2

b = np.linspace(0, 5, num=100)
h = 2
S = 4

q = np.linspace(0, 1, num=100)
W = np.zeros(100)

for i in range(100):
    W[i] = np.sum(f(b[b >= q[i]*S]) * (b[b >= q[i]*S] - h)) - C(q[i])

fig, ax = plt.subplots()
ax.plot(q, W, color='black')
ax.set(xlabel=r'$p$', ylabel=r'$W$',
       title= r'$W(q, S)$ avec $f(b) =\exp(-b)$')
ax.grid()
plt.savefig('Welf_h_seuil_exp.png', dpi=300)
plt.show()
```

Cependant, une fois que la probabilité de détection atteint un certain seuil, le bien-être commence à décroître. Cela s'explique par le fait que, lorsque la probabilité de détection devient très élevée, le coût de détection devient très élevé, ce qui réduit le bien-être global.

```python
def f(b):
    return 1 / (1 + np.exp(-b))

def C(q):
    return q**2

b = np.linspace(0, 5, num=100)
h = 2
S = 4

q = np.linspace(0, 1, num=100)
W = np.zeros(100)

for i in range(100):
    W[i] = np.sum(f(b[b >= q[i]*S]) * (b[b >= q[i]*S] - h)) - C(q[i])

fig, ax = plt.subplots()
ax.plot(q, W, color='black')
ax.set(xlabel=r'$q$', ylabel=r'$W$',
       title=r'$W(q, \bar{S})$ avec $f(b) =\frac{1}{1 + \exp(-b)}$')
ax.grid()
plt.savefig('Welf_f_logistique.png', dpi=300)

plt.show()
```

Enfin, le coût social associé à la dissuasion du crime est décrit par la fonction $C(q)$, qui dépend du niveau de dissuasion du crime mesuré par la probabilité $q$ de détection d'un criminel.

L'interprétation générale de cette courbe est que pour des niveaux de dissuasion élevés (c'est-à-dire pour des valeurs de $q$ élevées), le bien-être social diminue en raison des coûts associés à la détection et à la poursuite de criminels qui augmentent avec le niveau de dissuasion. Pour des niveaux de dissuasion faibles (c'est-à-dire pour des valeurs de $p$ faibles), le bien-être social est également faible en raison du manque de dissuasion du crime. # Intégrer la réputation dans le modèle

```python
import numpy as np
import matplotlib.pyplot as plt

# Fonction de densité de bénéfice
def f(b):
    return np.exp(-b)

# Coût de la détection en fonction de la probabilité de détection p
def C_p(p):
    return p**2

# Coût marginal de punition
def C_q(q):
    return q**2

# Bénéfice marginal monétaire
def benefit_monetary(q, S, h):
    return S*(q*S - h)*f(q*S)

# Bénéfice marginal de réputation
def benefit_reputation(r_bar, q, r_tilde, g):
    return -r_bar*(1 - q)*r_tilde*g(r_tilde)

# Fonction de la réputation
def g(r_tilde):
    return np.exp(-r_tilde)

# Fonction de coût total
def C_total(p, q, S, h, r_bar, r_tilde):
    return C_p(p) + C_q(q) - benefit_monetary(q, S, h) - benefit_reputation(r_bar, p, r_tilde, g)

# Optimisation du bien-être
def optimize_welfare(p_range, q_range, S, h, r_bar, r_tilde):
    W = np.zeros((100, 100))
    for i, p in enumerate(p_range):
        for j, q in enumerate(q_range):
            W[i, j] = C_total(p, q, S, h, r_bar, r_tilde)
    i_optimal, j_optimal = np.unravel_index(np.argmin(W), W.shape)
    p_optimal = p_range[i_optimal]
    q_optimal = q_range[j_optimal]
    return p_optimal, q_optimal

# Paramètres
p_range = np.linspace(0, 1, num=100)
q_range = np.linspace(0, 1, num=100)
S = 5
h = 1
r_bar = 1
r_tilde = 0.5

# Optimisation du bien-être
p_optimal, q_optimal = optimize_welfare(p_range, q_range, S, h, r_bar, r_tilde)

# Tracer les graphiques
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
ax[0].plot(p_range, [C_total(p, q_optimal, S, h, r_bar, r_tilde) 
for p in p_range], color='black', label='Cost vs. Probability of Arrestation')

#Plotting the cost vs. probability of punishment

ax[1].plot(q_range, [C_total(p_optimal, q, S, h, r_bar, r_tilde) for q in q_range], color='red', label='Cost vs. Probability of Punishment')
ax[1].set_xlabel('Probability of Punishment')
ax[1].set_ylabel('Cost')
ax[1].set_title('Cost vs. Probability of Punishment')
ax[1].legend()

plt.show()
```
